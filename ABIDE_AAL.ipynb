```python
base_dir = "/media/shuo/MyDrive/data/brain/abide/"
fmri_data_path = "ABIDE_pcp/cpac/filt_global/"
pheno_fname = "ABIDE_pcp/Phenotypic_V1_0b_preprocessed1.csv"
```

```python
pheno_file = os.path.join(base_dir, pheno_fname)
pheno_info = pd.read_csv(pheno_file, index_col=0)
```

```python
pheno_info
```

```python
data_left = []
data_right = []
idx_list = []
data_dir = os.path.join(base_dir, fmri_data_path)
for idx in pheno_info.index.values:
    data_fname = "%s_rois_aal.1D" % pheno_info.loc[idx, "FILE_ID"]
    data_fpath = os.path.join(data_dir, data_fname)
#     print(data_fpath)
    if os.path.exists(data_fpath):
        data = np.loadtxt(data_fpath, skiprows=0)[:,:-8]
        data_left.append(data[:, 0::2])
        data_right.append(data[:, 1::2])
        idx_list.append(idx)
```

```python
pheno_aal = pheno_info.loc[idx_list]
```

```python
pheno_aal.to_csv(os.path.join(base_dir, "Phenotype_filt_noqc.csv"), index=False)
```

```python
from nilearn.connectome import ConnectivityMeasure

correlation_measure = ConnectivityMeasure(kind='correlation', vectorize=True, discard_diagonal=True)
func_net_left = correlation_measure.fit_transform(data_left)

correlation_measure = ConnectivityMeasure(kind='correlation', vectorize=True, discard_diagonal=True)
func_net_right = correlation_measure.fit_transform(data_right)
```

```python
from scipy.io import savemat

savemat("abide_intra_half.mat", {"Left": func_net_left, "Rigit": func_net_right})
```
