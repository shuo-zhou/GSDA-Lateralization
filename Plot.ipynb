```python
import matplotlib.pylab as plt
%matplotlib inline
import numpy as np
```

```python
n_basis = [2, 5, 10, 15]
pc_com_male = [0.9263157894736842, 0.9978947368421054, 0.9968421052631579, 0.9947368421052631]
pc_com_female = [0.9304347826086957, 0.9991304347826088, 0.9947826086956523, 0.9991304347826088]
pc_ind_male = [0.6557894736842106, 0.43157894736842106, 0.4431578947368421, 0.44526315789473686]
pc_ind_female = [0.5695652173913043, 0.43478260869565216, 0.3956521739130435, 0.36]
```
